```python
class Dog:
    DogCount = 0
    def __init__(self,name,size,age,colour):
        self.name = name
        self.size = size
        self.age=age
        self.colour=colour
    @classmethod
    def createdog(cls, name, size, age, colour):
        cls.DogCout = cls.DogCount + 1
        return cls(name, size , age, colour)
    @classmethod
    def get_dog_count(cls):
        return cls.DogCount
obj = Dog.createdog("bull","large")    

```
