```python
class Dog:
    DogCount = 0
    species = "Dog"

    def __init__(self, name, size, age, color):
        self.name = name
        self.size = size
        self.age = age
        self.color = color

    @classmethod
    def create_dog(cls, name, size, age, color):
        cls.DogCount += 1
        return cls(name, size, age, color)

    @classmethod
    def get_dog_count(cls):
        return cls.DogCount


obj1 = Dog.create_dog("Buddy", "Medium", 5, "Brown")
obj2 = Dog.create_dog("Max", "Small", 3, "Black")

print("obj1 is a {}".format(obj1.__class__.species))
print("obj2 is also a {}".format(obj2.__class__.species))

print("{} is {} years old".format(obj1.name, obj1.age))
print("{} is {} years old".format(obj2.name, obj2.age))

print("Total Dogs:", Dog.get_dog_count())
```

# Ví dụ 1.2.2.1

- Class Dog gồm thuộc tính lớp `DogCount`, `species`
- Thuộc tính đối tượng gồm: name, size, age, color
- Sử dụng `@classmethod` để tạo đối tượng và đếm số lượng
