```python
class Dog:
    """
    Minh họa lớp Dog với các thuộc tính protected
    """
    _DogCount = 0   # protected: thuộc tính LỚP (giao diện protected)

    def __init__(self, name: str, size: str, age: int):
        self.name = name   # protected: thuộc tính protected (theo quy ước)
        self.size = size
        self.age = age

        Dog._DogCount += 1   # tăng thuộc tính lớp

    def display(self):  # Phương thức khong protected
        print(f"Name is a {self._name}")
d = Dog("Bull", "large", 2)

d._name   # ❌ Lỗi: không truy cập được do thuộc tính _name là protected
```

```python
d = Dog("Bull", "large", 2)

d.display()      # ✅ Phương thức protected: dùng trong lớp
print(d._name)    # ✅ Thuộc tính protected: dùng trong lớp

d.name            # ❌ Lỗi không truy cập được do thuộc tính 'name' không tồn tại
```
